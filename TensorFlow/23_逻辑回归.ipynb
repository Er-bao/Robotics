#  逻辑回归误差函数（交叉熵）梯度下降的推导
$Guess=\dfrac{1}{1+e^{-\left(m*x+b\right)}}$  
$CrossEntropy=-((Actual)*log(Guess)+(1-Actual)*log(1-Guess))$  
$误差函数(m,b)=-((Actual)*log(\dfrac{1}{1+e^{-\left(m*x+b\right)}})+(1-Actual)*log(1-\dfrac{1}{1+e^{-\left(m*x+b\right)}}))$

# 分别求误差函数对m和b的偏导数

```python
from sympy import * 
Actual_i,m,x_i,b,n=symbols('Actual_i,m,x_i,b,n')
errfunction=-Sum(Actual_i*log(1/(1+exp(-(m*x_i+b))))+(1-Actual_i)*log(1-1/(1+exp(-(m*x_i+b)))),(x_i,1,n))
```

```python
init_printing(pretty_print=True) #初始化latex显示
errfunction
```

```python
diff(errfunction,m)
```

```python
diff(errfunction,b)
```

# 误差函数交叉熵对m的偏导

```python
simplify(diff(errfunction,m))
```

# 误差函数交叉熵对b的偏导

```python
simplify(diff(errfunction,b))
```
