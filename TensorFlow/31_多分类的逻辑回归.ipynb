# 计算看电视的二分类逻辑回归

```python
import numpy as np
data=np.array([
    [5,1],
    [15,1],
    [25,0],
    [35,0],
    [45,0],
    [55,0],
    [65,0]    
])
feature=data[:,0:1]
label=data[:,-1:]
Feature=np.hstack((feature,np.ones((len(data),1))))
Feature
```

```python
Weight=np.ones((2,1))
```

```python
#记录mse对m和b变化率的惯性
m=np.zeros((2,1))
#记录mse对m和b变化率的速度
v=np.zeros((2,1))
learningrate = 0.1
def sigmoid(z):
    return 1/(1+np.exp(-z))

def gradentdecent():
    global Weight,m,v
    #计算的是m和b的梯度
    slop=np.dot(Feature.T,(sigmoid(np.dot(Feature,Weight))-label))
    #核心逻辑，Adam    
    beta_1=0.9
    beta_2=0.999    
    m=beta_1*m+(1-beta_1)*slop
    v=beta_2*v+(1-beta_2)*(slop**2)    
    m_p=m/(1-beta_1)
    v_p=v/(1-beta_2)    
    Weight=Weight-learningrate*m_p/np.sqrt(v_p+0.000000001)
    return slop
```

```python
for i in range(10000):
    slop=gradentdecent()
    if np.sum(np.square(slop))<0.01:
        print("i={},训练完毕后的Weight：{},slop={}".format(i,Weight,slop))
        break
```

```python
y=[]
x=np.linspace(1,100,100)
for i in range(len(x)):
    y.append(sigmoid(np.dot(np.array([i,1]),Weight)))
    
import matplotlib.pyplot as plt
%matplotlib notebook
plt.plot(x,y)
plt.show()
```
