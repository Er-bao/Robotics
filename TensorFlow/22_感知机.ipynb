# 完整的感知机的实现

```python
# 声明激活函数
def stepFunction(value):
    if value>=0:
        return 1
    else:
        return 0
# 声明感知机的运算
# X数据矩阵（x1,x2,...）
# W权重矩阵（w1,w2,...）
def preception(X,W,b):
    return stepFunction((np.matmul(X,W)+b)[0])#0,1     X和W向量相乘
```

```python
#训练感知机
#X 原始的数据feature
#y 原始真实的结果
#b 偏置量
#W 权重矩阵（w1,w2,...）
def trainPreception(X,y,W,b,learningrate=0.01):
    for i in range(len(X)):
        y_guess=preception(X[i],W,b)
        
        ## 如果真实是苹果，预测是香蕉，往下移动线
        if y[i]-y_guess==1:#1-0
            W[0]+=learningrate*X[i][0]
            W[1]+=learningrate*X[i][1]
            b+=learningrate
            
        ## 如果真实是香蕉，预测是苹果，往上移动线
        if y[i]-y_guess==-1:#0-1
            W[0]-=learningrate*X[i][0]
            W[1]-=learningrate*X[i][1]
            b+=learningrate
    
    return (W,b)
```

```python
## 多次调用trainPrection让线收敛到合适的位置

#W=np.array(np.random.rand(2,1))
#b=np.random.rand(1)[0]
np.random.seed(42)
W=np.random.rand(2,1)#m1,m2随机得到两个权重，x1和x2的系数
b=-1
W,b
```

```python
x=data[:,0:2]
y=data[:,-1]
#for i in range(len(x)):
 #   print(preception(x[i],W,b))
```

```python
#存储所有的历史线的方程
history_lines=[(-W[0][0]/W[1][0],-b/W[1][0])]
for i in range(50):
    W,b=trainPreception(x,y,W,b)
    # W[0]*x1+W[1]*x2+b=0
    # x2=-W[0]/W[1]*x1-b/W[1]
    history_lines.append((-W[0][0]/W[1][0],-b/W[1][0]))
```
